# AIES

```python

# Load saved samples
data = np.load("mcmc_samples_AIES.npz")
samples = {k: data[k] for k in data.files}

# Convert to ArviZ InferenceData
idata = az.from_dict(posterior=samples)

# Plot and summarize
az.plot_trace(idata)
print(az.summary(idata))
corner_plot = corner.corner(data = {'m': samples['m_tot'], 
                               'r_s': samples['r_s'], },)
```
