```python
from galpy.potential import MWPotential2014, HernquistPotential
from galpy.util.conversion import mass_in_msol

import numpy as np
import astropy.units as u
```

```python
MW_PSP, MW_MN, MW_NFW = MWPotential2014

ro = 8.0 # kpc
vo = 220.0 # km/s

mass_conv = mass_in_msol(vo, ro)

print("Bulge (PSP)", vars(MW_PSP))
print("Disk (MN)", vars(MW_MN))
print("Halo (NFW)", vars(MW_NFW))
```

```python
# Bulge: PowerSphericalPotentialwCutoff
alpha = MW_PSP.alpha
r_c   = MW_PSP.rc * ro
M_bulge = MW_PSP.mass(np.inf, use_physical=True, ro=ro, vo=vo)

print(f"Bulge (PSP):  alpha = {alpha}")
print(f"              r_c   = {r_c} kpc")
print(f"              M = {M_bulge} M☉")

# Disk: MiyamotoNagaiPotential
a_disk = MW_MN._a * ro
b_disk = MW_MN._b * ro
M_disk = MW_MN._amp * mass_conv

print(f"\nDisk (MN):    a = {a_disk} kpc")
print(f"              b = {b_disk} kpc")
print(f"              M = {M_disk} M☉")

# Halo: NFWPotential
r_s = MW_NFW.a * ro
r_vir = MW_NFW.rvir(use_physical=True, ro=ro, vo=vo)
M_char = MW_NFW._amp * mass_conv
M_vir = MW_NFW.mass(r_vir, use_physical=True, ro=ro, vo=vo)

print(f"\nHalo (NFW):   a (r_s) = {r_s} kpc")
print(f"              r_vir = {r_vir} kpc") 
print(f"              M_char (4π rho_s r_s³) = {M_char} M☉")
print(f"              M_vir = {M_vir} M☉")
```

```python
# LMC: HernquistPotential
LMC = HernquistPotential(amp=2e11 * u.Msun, a=5. * u.kpc/(1. + np.sqrt(2.)))
print(" ", vars(LMC))
```

```python
r_s_LMC = LMC.a * ro
M_char_LMC = LMC.mass(np.inf, use_physical=True, ro=ro, vo=vo)

print(f"\nLMC (Hernquist): a = {r_s_LMC} kpc")
print(f"                 M = {M_char_LMC} M☉")
```
