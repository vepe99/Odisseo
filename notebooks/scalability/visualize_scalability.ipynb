```python
import numpy as np
import matplotlib.pyplot as plt

runtime_list_direct_acc =  np.load('./runtime_list_direct_acc.npy')
runtime_list_matrix =  np.load('./runtime_list_matrix.npy')
runtime_list_laxmap = np.load('./runtime_list_laxmap.npy')

N_particles_list = [10, 100, 500, 1_000, 10_000, 100_000]

# Extract mean and std from runtime_list tuples
mean_times_acc = [t[0] for t in runtime_list_direct_acc]
std_times_acc = [t[1] for t in runtime_list_direct_acc]

mean_times_acc_matrix = [t[0] for t in runtime_list_matrix]
std_times_acc_matrix = [t[1] for t in runtime_list_matrix]

mean_time_acc_laxmap = [t[0] for t in runtime_list_laxmap]
std_time_acc_laxmap = [t[1] for t in runtime_list_laxmap]   

fig = plt.figure(figsize=(15, 5), tight_layout=True)
ax = fig.add_subplot(122)
# Create an error bar plot
ax.errorbar(N_particles_list, mean_times_acc, yerr=std_times_acc, fmt='o', capsize=5, 
             markersize=8, ecolor='red', markeredgecolor='black', markerfacecolor='blue', label='DIRECT ACC')
ax.errorbar(N_particles_list, mean_times_acc_matrix, yerr=std_times_acc_matrix, fmt='o', capsize=5,
                markersize=8, ecolor='red', markeredgecolor='black', markerfacecolor='green', label='MATRIX')
ax.errorbar(N_particles_list, mean_time_acc_laxmap, yerr=std_time_acc_laxmap, fmt='o', capsize=5,
                markersize=8, ecolor='red', markeredgecolor='black', markerfacecolor='red', label='LAXMAP')

# Set scales and labels
ax.set_xscale('log')
# ax.set_yscale('log')
ax.set_xlabel('N particles')
# ax.set_ylabel('Runtime [s]')
ax.grid(True, which="both", ls="--", alpha=0.7)
ax.legend()


ax = fig.add_subplot(121)
# Create an error bar plot
ax.errorbar(N_particles_list[:4], mean_times_acc[:4], yerr=std_times_acc[:4], fmt='o', capsize=5, 
             markersize=8, ecolor='red', markeredgecolor='black', markerfacecolor='blue', label='DIRECT ACC')
ax.errorbar(N_particles_list[:4], mean_times_acc_matrix[:4], yerr=std_times_acc_matrix[:4], fmt='o', capsize=5,
                markersize=8, ecolor='red', markeredgecolor='black', markerfacecolor='green', label='DIRECT ACC MASKING')
ax.errorbar(N_particles_list[:4], mean_time_acc_laxmap[:4], yerr=std_time_acc_laxmap[:4], fmt='o', capsize=5,
                markersize=8, ecolor='red', markeredgecolor='black', markerfacecolor='red', label='LAXMAP')

# Set scales and labels
# ax.set_xscale('log')
ax.set_xlabel('N particles')
ax.set_ylabel('Runtime [s]')
ax.grid(True, which="both", ls="--", alpha=0.7)
ax.legend()

# fig.savefig('runtime_direct_acc_vs_direct_acc_matrix.pdf')
```
